```python
#import libraries
import qrcode
import matplotlib.pyplot as plt
```

```python
img = qrcode.make("RCCG Christ Chapel Youth Student Information Session click the link please")
```

```python
data = "https://docs.google.com/forms/d/e/1FAIpQLScp7BtZ4UdsMQcC_HNLrJIhp7P3G8Vt6hwdmzXeDlJlUT4jrw/viewform"
```

```python
#Output file name
filename = "My_QR_CODE.png"
```

```python
qr = qrcode. QRCode(version= 1,
                   error_correction=qrcode.constants.ERROR_CORRECT_L,
                   box_size = 50,
                   border=20)
```

```python
#generate the qr code

image = qrcode.make(data)
```

```python
image.save(filename)
plt.imshow(image,cmap="gray")
```
